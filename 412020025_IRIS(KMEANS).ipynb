```python
import pandas as pd
data = pd.read_csv('/content/gdrive/My Drive/Dataset/Iris.csv')
```

```python
data.head()
```

```python
label = data[['Species']]
new_data = data.drop(columns = ["Id", "Species"])
data = data.drop(columns = ["Id"])
```

```python
import numpy as np
labell = np.array(label)
labell
```

```python
new_data
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 3)
km.fit(new_data)
```

```python
inertia = []
for i in range(1,11):
  km_test = KMeans(n_clusters = i).fit(new_data)
  inertia.append(km_test.inertia_)
```

```python
range(1,11)
```

```python
import seaborn as sns
sns.lineplot([x for x in range(1,11)], inertia).set(xlabel = "Iteration", ylabel = "Inertia")
```

```python
new_labels = km.labels_
new_labels = new_labels.astype(np.dtype(object))
new_labels
```

```python
new_data["KMeans Species"] = new_labels
```

```python
new_labels = new_data["KMeans Species"].to_numpy()
new_labels
```
